```python
import numpy as np
import pylab as plt

scan_obj=data_dict[0]
left_lidar=scan_obj.left
right_lidar=scan_obj.right

L_x=right_lidar['x']

holder=np.fft.fft(L_x)

frequencies=np.fft.fftfreq(len(holder))

# print frequencies

plt.figure(0)
plt.plot(frequencies, np.abs(holder)) 
plt.show()
```
